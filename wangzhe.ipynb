```python
import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg 
import numpy as np
 
wangzhe = mpimg.imread('wangzhe.jpg') 
luban = mpimg.imread('luban.jpg')
dianwei = mpimg.imread('dianwei.jpg')
zhaoyun = mpimg.imread('zhaoyun.jpg')
jiazai = mpimg.imread('jiazai.jpg')

plt.imshow(wangzhe) # 显示图片
plt.axis('off') # 不显示坐标轴
plt.show()
class Wangzhe(object):
    def __init__(self):
        pass
    def duizhan(self):
        print('请选择人机或者多人对战！')
        res = input('请输入>>')
        print("您选的是：",res)
              
        
    def renwu(self):
        print('请选择英雄：典韦、赵云、鲁班')
        ren_ = input('请输入>>')
        print("您选的是：",ren_)
        if self.renwu == '典韦':
            plt.imshow(dianwei) 
            plt.axis('off') 
            plt.show()
            print("您选择的典韦：战力—1568，防御—5600")
        elif self.renwu == '赵云':
            plt.imshow(zhaoyun) 
            plt.axis('off') 
            plt.show()
            print("您选择的赵云：战力—2345，防御—2688")
        else:
            plt.imshow(luban) 
            plt.axis('off') 
            plt.show()
            print("您选择的鲁班：战力—1356，防御—3579")
    def suiji(self):
        print("系统将为您随机产生对战玩家")
        res = np.random.choice(['典韦','赵云','鲁班'])
        print('您的对手为>>',res)
    def kaishi(self):
        print("请输入开始")
        KS= input('输入>>')
        if  KS== "开始":
            print("输入正确")
        else:
            print("输入有误！")
    def jiazai(self):
        print('进入加载页面 ...')
        plt.imshow(luban) 
        plt.axis('off') 
        plt.show()
joker = Wangzhe()
joker.duizhan()
joker.renwu()
joker.suiji()

joker.kaishi()
joker.jiazai()
```
